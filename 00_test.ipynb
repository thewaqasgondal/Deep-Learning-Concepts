### Connecting Colab to GitHub

To save your Colab notebook to a GitHub repository, follow these steps:

1.  **Ensure you are logged into your GitHub account.**
2.  In Google Colab, go to `File` in the top menu.
3.  Select `Save a copy to GitHub...`.
4.  A dialog box will appear. You will need to:
    *   Select your GitHub repository from the dropdown menu (e.g., `thewaqasgondal/Deep-Learning-Concepts`).
    *   Choose the branch you want to save to (usually `main` or `master`).
    *   Provide a commit message describing your changes.
    *   You can also specify the filename if you want to change it.
5.  Click `OK` to save the notebook to your repository.

### Real-time Synchronization

Colab does not offer a direct 'real-time' two-way synchronization like a local Git client. However, once you've saved a copy to GitHub:

*   **To push new changes:** Repeat the `File` > `Save a copy to GitHub...` process. Colab will detect that the file already exists and will allow you to commit updates.
*   **To pull changes from GitHub:** If you or someone else has updated the notebook directly on GitHub, you'll need to reopen the notebook from GitHub in Colab (`File` > `Open notebook` > `GitHub` tab) to get the latest version. You might want to save your current work first if you have unsaved changes locally.

This process allows you to manage your notebook versions within your GitHub repository.
